```python
df=pd.concat([train_new,test],axis=0)
df
```

```python
X,y=df.iloc[:,:-1],df.iloc[:, [-1]]
y=y.replace({"Team1":1 , "Team2":2})
```

```python
#performing pca
from sklearn.decomposition import PCA
pca= PCA(n_components=100)
X_princple_components=pca.fit_transform(X)
X_princple_components
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
X_princple_components=scaler.fit_transform(X_princple_components)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test, y_train , y_test = train_test_split(X_princple_components,y, test_size=0.20,random_state=44)
```

```python
param_grid = {'n_estimators': [100, 300, 500, 800, 1200],
              'max_depth': [5, 10, 15, 20, 25, 30, None],
              'min_samples_split': [2, 5, 10, 15, 100],
              'min_samples_leaf': [1, 2, 4, 10],
              'max_features': ['sqrt', 'log2', None]}
```

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestRegressor(random_state=42)

from sklearn.model_selection import RandomizedSearchCV

rf_random = RandomizedSearchCV(estimator=rf, param_distributions=param_grid, n_iter=100, cv=5, random_state=42, n_jobs=-1)

```

```python
rf_random.fit(X_train, y_train)
```

```python
print(rf_random.best_params_)
```

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(n_estimators= 500, min_samples_split= 15, min_samples_leaf= 10, max_features= None, max_depth= 20)
```

```python
rf.fit(X_train,y_train)
```

```python
y_pred = rf.predict(X_test)
y_pred=pd.DataFrame(y_pred)
y_pred
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,f1_score, precision_score, recall_score
```

```python
accuracy_score(y_test, y_pred)
```

```python
recall_score(y_test, y_pred)
```

```python
precision_score(y_test, y_pred)
```

```python
f1_score(y_test, y_pred)
```

```python
confusion_matrix(y_test,y_pred)
```
